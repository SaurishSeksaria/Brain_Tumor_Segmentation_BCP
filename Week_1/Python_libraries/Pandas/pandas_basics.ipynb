```python
# TODO
df = pd.read_csv("videogamesales.csv")
df["global_sales"] = df["NA_Sales"]+df["EU_Sales"]+df["JP_Sales"]+df["Other_Sales"]
df = df.sort_values(by = "global_sales", ascending = False)
plotdf = df.groupby(by = "Genre")["global_sales"].sum()
#plotdf.head()
plotdf.plot(kind = "bar", title = "Genre Distribution")
plt.show()
#df.head()
df2 = df[df["Name"].str.contains("Grand Theft Auto")]
#df2.head()
df2["Total_Sales"] = df2["EU_Sales"]+df2["JP_Sales"]
answerdf = df2[["Name","Platform","Year","Total_Sales"]]
print(answerdf.head())
df2["global_sales"].plot(kind="pie")
plt.show()
```
